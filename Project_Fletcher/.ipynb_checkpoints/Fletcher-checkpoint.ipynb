### Pulling List of Streamers and their Broadcasts/Video IDS

```python
# Retrieve List of Broadcasts First

f = FloatProgress(min=0, max=len(list_of_channels))
display(f)
f.value = 0
for channel in list_of_channels['Channel_Title']:
    try:
        limit = 100
        offset = 0
        response = urllib.request.urlopen("https://api.twitch.tv/kraken/channels/{}/videos?broadcasts=true&limit={}&offset={}".format(channel,limit,offset))
        json_data = simplejson.load(response, encoding='UTF-8')
        filename = channel+"_broadcasts"+".json"
        with open(filename, 'w') as outfile:
            json.dump(json_data, outfile, sort_keys = True, indent = 4, ensure_ascii=False)
    except:
        print("Failed to grab data from :" + channel)
#     print("Done with :", channel)
    f.value = f.value+1
!mv *.json TwitchJSONFiles
print("All Done with pulling list of broadcasts!")
```

## Placing Chats into a DataFrame

```python
fake_array = []

top_directory = '/Users/Hayw1re/ds/metis/classwork/sf16_ds3/projects/floofydugong/TwitchJSONFiles'
SPLIT_SENTENCES = re.compile(u"(\d.+\))\W{2}\<(.+)\>\W{1}(.+)")
for root, dirs, files in os.walk(top_directory):
    streamer_name = (root.split('/')[-1])
    for fname in filter(lambda fname: fname.endswith('.txt'), files):
        document = open(os.path.join(root, fname)).read()
        for sentence in SPLIT_SENTENCES.findall(document):
            fake_array.append([streamer_name, fname[0:-4],sentence[0],sentence[1],sentence[2]])
fake_array
answer = pd.DataFrame(fake_array)
answer.head(10)
```

```python
def convert_time(time_string):
    try:
        test_string = dateutil.parser.parse(time_string)
        return test_string.time()
        print("nice!")
    except:
        return ("failed at :",time_string)
```

```python
# pogchamp in or out?
stop = stopwords.words('english')
stop += ['.', ',', '(', ')', "'", '"', '', 'pogchamp']
```

###  Flow : Countvectorizer > LSA

```python
num_of_comps = 10
# Run countvectorizer
vectorizer = CountVectorizer(min_df = 1, stop_words = stop)

#Creates a spare matrix
dtm = vectorizer.fit_transform(dota2ti_game_one['Comment']) 

# Makes this into a df to read/review
df = pd.DataFrame(dtm.toarray(), index=dota2ti_game_one['Comment'], columns=vectorizer.get_feature_names())
df

#randomized vs arpack
lsa = TruncatedSVD(n_components=num_of_comps, algorithm = 'randomized')
dtm_lsa = lsa.fit_transform(dtm)
# dtm_lsa = Normalizer(copy=False).fit_transform(dtm_lsa)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x=range(num_of_comps),y=lsa.explained_variance_ratio_)
plt.show()
```

```python
num_of_components = ["component_"+str(i) for i in range(num_of_comps)]

# print(num_of_components)

review_answer = pd.DataFrame(lsa.components_.round(5),index = [num_of_components],columns = vectorizer.get_feature_names())

for index, val in enumerate(num_of_components):
    word_and_variance = pd.DataFrame(review_answer.iloc[index,:]).sort_values(by=val,ascending=False).head(10)
    pprint(word_and_variance)
```

###  Flow : TFIDF > LSA

```python
num_of_comps = 5

tfidf_vectorizer = TfidfVectorizer(min_df = 1, stop_words = stop, ngram_range=(2,3))
tfidf_dtm = tfidf_vectorizer.fit_transform(dota2ti_game_one['Comment'])
# pd.DataFrame(tfidf, columns=vectorizer.get_feature_names())

# tfidf_dtm.toarray()

df = pd.DataFrame(tfidf_dtm.toarray(), index=dota2ti_game_one['Comment'], columns=tfidf_vectorizer.get_feature_names())
df.head()
```

```python
tfidf_lsa = TruncatedSVD(n_components=num_of_comps, algorithm = 'randomized')
dtm_lsa = tfidf_lsa.fit_transform(tfidf_dtm)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x=range(num_of_comps),y=tfidf_lsa.explained_variance_ratio_)
plt.show()
```

```python
num_of_components = ["component_"+str(i) for i in range(num_of_comps)]

# print(num_of_components)

review_answer = pd.DataFrame(tfidf_lsa.components_.round(5),index = [num_of_components],columns = tfidf_vectorizer.get_feature_names())

for index, val in enumerate(num_of_components):
    word_and_variance = pd.DataFrame(review_answer.iloc[index,:]).sort_values(by=val,ascending=False).head(10)
    pprint(word_and_variance)
```
