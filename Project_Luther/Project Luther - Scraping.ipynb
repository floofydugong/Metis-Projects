## Retrieving Max Pages

**Description:** This will identify the maximum value of entries. From that, I divide by the maximum number of entries possible for a page to determine how many pages to loop through.

```python
url = 'https://www.rottentomatoes.com/critic/ao-scott/movies?cats=&genreid=&letter=&switches=&sortby=&limit=50&page=1'
response = requests.get(url)
page = response.text
soup = BeautifulSoup(page, 'lxml')

number_of_entries = soup.find(class_="clearfix default-half-padding").text
max_entries = number_of_entries.split()[-1]

total_pages = int(max_entries)/50
max_page = int(np.ceil(total_pages))
print("Done calculating. Max pages = {}".format(max_page))
```

```python
# Create a list of urls to iterate through after finding the maximum number of pages
list_of_urls = []

for i in range(1,max_page+1):
    url = 'https://www.rottentomatoes.com/critic/ao-scott/movies?cats=&genreid=&letter=&switches=&sortby=&limit=50&page='+str(i)
    list_of_urls.append(url)
print("Done creating {} URLs".format(max_page))
```

## RT Scraping Part I (Title, Release Year, Critic Score, Movie URL)

**Description:** This code will iterate through all of the urls previously generated to retrieve the movie data. The intiial strucutre I'm viewing is a list of all the movies that A.O Scott has critiqued, so there are only a few columns worth of information I'm going to grab. I grab data column by column (versus row by row) and then concatenate the lists at the end into a dataframe. For the Genre and runetime, I will be using the URL retrieved from the first pull to grab these two pieces of information.

**Scrape Breakdown**
1. Movie Title - Title of the Movie
2. Release Date - Release Data as a Year
3. tMeterSCore - The score given by certified critics (out of 100)
4. ao_scott_rot_fresh - A "fresh" or "rotten" label given by A.O Scott, stored as a 1 or 0 respectively
5. movie_url - The RT tomato URL used to access the movie's specific page.

**Notes:** As a safety measure, I also save the data into a CSV for easy access/retrieval. I also have print statements to let me know where I am in the scraping process

```python
movie_title = []
release_date= []
tMeterScore = []
ao_scott_rot_fresh = []
movie_url = []

counter = 1
for link in list_of_urls:
    response = requests.get(link)
    page = response.text
    soup = BeautifulSoup(page, 'lxml')

    #Movie Title / URL
    title_data = soup.find_all('a', class_="unstyled articleLink",target="_top")
    for title in title_data:
        movie_url.append(title['href'])
        movie_title.append(title.text)
        
    # Release Date
    date_data = soup.find_all('td',class_="center")
    for date in date_data:
        try:
            release_date.append(re.findall(r'(?<=\()\d{4}(?=\))', date.text))
        except:
            release_date.append(np.nan)

    # tMeterScore
    re_tmeter = re.compile(r'noRating|tMeterScore')
    tMeterScore_data = soup.find_all('span',class_=re_tmeter)
    for score in tMeterScore_data:
        if score.text == ' No Score Yet ':
            tMeterScore.append(np.nan)
        else:
            tMeterScore.append(score.text)

    # ao_scott_data
#     re_rot_fresh = re.compile(r'rotten$|fresh$')
#     re_division = re.compile(r'[^a-zA-Z]+')
    
    find_table = soup.find("table",class_="table table-striped").find_all("tr")
    
    for item in find_table:
        first_col_tag = item.find("td")
        if first_col_tag:
            span_tag = first_col_tag.find("span")
            if span_tag:
                if "fresh" in span_tag.get("class"):
                    ao_scott_rot_fresh.append(1)
                else:
                    ao_scott_rot_fresh.append(0)

    time.sleep(2)
    print ('Completed {} run(s)'.format(counter))
    counter += 1

# Merging Dataframes together
title_df = pd.DataFrame(movie_title, columns=['Title'])
date_df = pd.DataFrame(release_date, columns=['ReleaseYear'])
score_df = pd.DataFrame(tMeterScore, columns=['CriticsTomatometer'])
url_df = pd.DataFrame(movie_url, columns=['MovieUrl'])
ao_df = pd.DataFrame(ao_scott_rot_fresh, columns=['AORotFresh'])

rott_tomato_df = pd.concat([title_df, date_df,score_df,url_df,ao_df], axis=1)

# CSV Safety Measure
rott_tomato_df.to_csv('rt_data.csv')

print("Completed Part I of scraping!")
```

## RT Scraping Part II (run time, genre)

**Description:** With the urls from the previous pull, I will then access each movie title and scrape the genre and run time. Following the same steps as before, I will be pulling the information into seperate lists before concating the data together as a dataframe, which will then be joined to the initially created dataframe from the first scrape.

**Notes:** One of the problems I ran into here was that the link provided by RT would return a 404. I'm not sure why faulty links were provided, but I circumvented the issue by adding in a "try" and "except" block. Like the previous code snippet, I also added in print statement to show the current status.

```python
rt_run_time_data = defaultdict(list)
rt_genre_data = defaultdict(list)
counter = 1

for index in range(0,len(rott_tomato_df)):
    title = rott_tomato_df.iloc[index]['Title']
    
    # Create the URL
    url_first_half = "https://www.rottentomatoes.com"
    url_second_half = (rott_tomato_df.iloc[index]['MovieUrl'])
    full_url = url_first_half + url_second_half
    
    # Fix Broken Links
    try:
        response = requests.get(full_url)
        page = response.text
        soup = BeautifulSoup(page, "lxml")
        f = open('workfile.txt', 'a')
        f.write("response status for {} : {}\n".format(full_url, str(response.status_code)))
    except requests.exceptions.TooManyRedirects:
        fixed_url = 'https://www.rottentomatoes.com{}{}'.format('/',full_url)
        response = requests.get(fixed_url)
        page = response.text
        soup = BeautifulSoup(page, "lxml")
        print("Hit an Exception for the movie after {}: ".format(full_url), response.status_code)
  
    # Finding run_times
    re_time = re.compile(r'P.+')
    run_time_list = soup.find("time", datetime=re_time)
    try:
        rt_run_time_data[title].append((run_time_list.text).strip())
    except:
        rt_run_time_data[title].append(np.NaN)
        
    genre_finds = soup.find_all('genre:link')
    for item in genre_finds:
        genre_finder = item.find("span")
        try:
            rt_genre_data[title].append(genre_finder.text)
        except:
            rt_genre_data[title].append(np.NaN)
    
    print ('Completed {} run(s)'.format(counter))
    counter += 1
    random_num = (np.random.randint(0,2))
    time.sleep(0+random_num)
f.close()
```

## Scraping from BoxOfficeMojo (Studio','LifetimeGross','LifetimeTheaters','OpeningGross','OpeningTheaters','ReleaseYear)

**Description:** For this scrape from BoxOfficeMojo, I essentially followed the same steps I used in retrieving the data from RT (using beautifulsoup). 

**Notes** One of the issues with the way that I pulled is that I only grabbed one movie from the search results, even if it had multiple. Generally, the most relevant movie would be at the top, but this was not always the case.

```python
bigger_list = []

# Loading page
for index in range(0,len(rott_tomato_df_both)):
    list_of_movies_bom = []
    title = rott_tomato_df_both.iloc[index]['Title']

    try:
        url = "http://www.boxofficemojo.com/search/?q={}".format(title)
        response = requests.get(url)
        page = response.text
        soup = BeautifulSoup(page, 'lxml')
        re_movies_found = re.compile(r'[0-9]+ Movie Matches')
        testing = soup.find("b",text=re_movies_found)
        num_movie_found = int((testing.text).split()[0])
        print("Found {} movie(s) for {}".format(num_movie_found,title))
        
        find_all_tables = soup.find_all("table")[4]
        cells = find_all_tables.findChildren('td')
        for cell in cells[6:14]:
            value = (cell.text).strip()
            list_of_movies_bom.append(value)
        bigger_list.append(list_of_movies_bom)
    except:
        print("No movies found for {}".format(title))
```

## Advanced Linear Regression / Modeling Analysis
